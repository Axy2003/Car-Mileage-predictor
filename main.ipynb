```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor 
import warnings
warnings.filterwarnings("ignore")
```

```python
URL = "Automobile.csv"
df = pd.read_csv(URL)

df.sample(10)
```

```python
print(df.shape)
df = df.dropna()
df.plot.scatter(x="displacement", y="mpg")
```

```python
target = df["mpg"]
features = df[["horsepower", "weight", "displacement"]]
```

```python
df_model = RandomForestRegressor()
df_model.fit(features, target)
```

```python
df_model.score(features, target)
```

```python
h = input("Horsepower:")
w = input("Weight:")
d = input("Displacement:")

df_model.predict([[h, w, d]])
```
